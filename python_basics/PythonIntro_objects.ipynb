# Objects

Python is an object oriented program, it is fundamentally based on objects, each having built in properties and methods

Why is the set of lists below cumbersome?

```python
kirk = ["James Kirk", 34, "Captain", "Enterprise"]
spock = ["Spock", 35, "Science Officer", "Enterprise"]
mccoy = ["Leonard McCoy", 40, "Chief Medical Officer", "Enterprise"]
```

```python
class Trekkie:
    universe = "Star Trek"

    def __init__(self, name, age, rank, ship):
        self.name = name
        self.age = age

    # Instance method
    def description(self):
        print(self.name +' is '+ str(self.age)+' years old')

    # Another instance method
    def reassign(self, newship):
        self.ship = newship
        print(self.name +' has been reassigned to the USS ' +self.ship)
    
    def addyear(self):
        self.age += 1
```

```python
kirk = Trekkie("James Kirk", 34, "Captain", "Enterprise")
kirk.description()
```

```python
print(kirk)
```

```python
print(kirk.name)
```

```python
print(kirk.reassign)
```

```python
kirk.reassign('Constellation')
```
