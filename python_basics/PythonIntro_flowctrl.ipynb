# If/Else

```python
a = 4
if a == 4:
    print('Yay a is 4')
else:
    print('Ugh a is not 4')
```

```python
a = 4
if a == 4:
    print('Yay a is 4')
if a != 4:
    print('Ugh a is not 4')
```

# For Loop

```python
for ind in [1,2,3,4]:
    print(ind*5)
```

What is the difference between the previous and this statement?

```python
for ind in range(4):
    print(ind*5)
```

## Looping over a string

```python
for val in 'Cassini':
    print(val)
```

But what if you also want the index?

Use enumerate!

```python
for ind, val in enumerate('Cassini'):
    print(str(ind)+' '+str(val))
```

```python
for ind,val in enumerate([1,2,3,4]):
    print(str(ind)+' '+str(val*5))
```

## Looping over a dictionary

```python
dict = {'Artemis':['Moon',2008,2],'Barrel':['Antarctica',2013,1],'Cassini':['Saturn',1997,1]}
print(dict.keys())
```

```python
dict = {'Artemis':['Moon',2008,2],'Barrel':['Antarctica',2013,1],'Cassini':['Saturn',1997,1]}
for mission in dict.keys():
    print(mission)
    print('Location: '+dict[mission][0])
    print('Launch: '+str(dict[mission][1]))
    print('Number of Spacecraft: '+str(dict[mission][2]))
    print('--------------------')
    print('\n')
```

You can also loop over multiple nested items at once

```python
lis = [[1,'a'],[2,'b'],[3,'c'],[4,'d']]
for num, let in lis:
    print(str(num)+' '+let)
```

# While Loop

Conditional

```python
val = 5
while val < 20:
    print('Value is: '+str(val))
    val += 2
```

```python
import datetime
import time
timeIn10 = datetime.datetime.now()+datetime.timedelta(seconds=10)
while datetime.datetime.now() < timeIn10:
    print('Waiting')
    time.sleep(1)
```

Be careful to not create infinite loops

Be careful to track the variable that has the conditional statement

# Try Except

Error Handling

```python
a = 5
b = 25
try:
    c = a/b
    print(c)
except ZeroDivisionError:
    print("Cannot Divide by 0")
except:
    print("Something else went wrong")
    
```

```python
a = 5
b = 0
try:
    c = a/b
    print(c)
except ZeroDivisionError:
    print("Cannot Divide by 0")
except:
    print("Something else went wrong")
    
```

Why is the above better than just letting the error flag?
