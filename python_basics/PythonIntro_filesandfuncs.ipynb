# Functions

```python
import datetime
def str2date(date_str):
    '''
    -
    #---------------
    # The "str2date" function converts a date string into a datetime
    #
    # Inputs:
    #   date_str - A date string in the format of 'YYYYMMDD' such as 20200601 for June 1, 2020 (string)
    #
    # Outputs
    #   date_dt - A datetime object of the date without hours, minutes, or seconds (datetime object)
    #
    # Errors
    #   Will print error and quit if 
    #       1. The date string is not 8 characters
    #       2. The date does not exist (ie the month is too big or the day of month is too big)
    #   Will print for:
    #       1. Success if complete
    #
    # Example:
    #   date_dt = str2date('20200601')
    #       -> 2020-06-01 00:00:00
    #       -> *** complete ***
    #
    #---------------
    -
    '''
    yr = int(date_str[:4])
    mo = int(date_str[4:6])
    dy = int(date_str[6:8])
    
    date_dt = datetime.datetime(yr,mo,dy)
    
    print(date_dt)
    print('*** complete ***')
    
    return date_dt
```

```python
date_dt = str2date('20200601')
```

# Files

```python
# Open the file
f = open('./Data/OrbitData.txt','r')
# Read from the file
raw = f.read()
# Close the File
f.close()
# Split the file by line breaks
dat = raw.split('\n')
# Print the Header
print(dat[0])
# Loop over every line checking for JUN 08
for ind, line in enumerate(dat):
    if 'JUN 08' in line:
        print(str(ind)+': '+line)
```
